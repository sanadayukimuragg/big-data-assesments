```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
integrated = pd.read_csv('studentInfo.csv')
integrated.shape
```

```python
integrated.head()
```

```python
integrated["code_module_presentation"]=integrated[['code_module','code_presentation']].agg('-'.join,axis=1)
integrated.head()
```

```python
integrated = integrated.astype({'id_student':'string'})
integrated["code_module_student_presentation"]=integrated[['code_module_presentation','id_student']].agg('-'.join,axis=1)
integrated.head()
```

```python
integrated['cohort'] = integrated['code_presentation'].str[-1:]
integrated.head()
```

```python
uniqueValues = integrated['imd_band'].unique()
print(uniqueValues)
```

```python
# replace "10-20" with "10-20%" 

integrated['imd_band'] = integrated['imd_band'].replace(['10-20'], '10-20%')
uniqueValues = integrated['imd_band'].unique()
print(uniqueValues)
```

```python
courses = pd.read_csv('courses.csv')
courses.head()
```

```python
courses["code_module_presentation"]=courses[['code_module','code_presentation']].agg('-'.join,axis=1)
courses = courses.drop(columns=['code_module','code_presentation'])
courses.head()
```

```python
# merge data into integrated dataframe

integrated = integrated.merge(courses,on="code_module_presentation", how="inner")
print(integrated.shape)
integrated.head()
```

```python
reg = pd.read_csv('studentRegistration.csv')
reg.shape
reg.head()
```

```python
reg = reg.astype({'id_student':'string'})
reg["code_module_student_presentation"]=reg[['code_module','code_presentation','id_student']].agg('-'.join,axis=1)
reg = reg.drop(columns=['code_module','code_presentation','id_student'])
reg.head()
```

```python
assmt = pd.read_csv('assessments.csv')
assmt.shape
```

```python
std_assmt = pd.read_csv('studentAssessment.csv')
std_assmt.shape
```

```python
merged_assmt = assmt.merge(std_assmt,on="id_assessment", how="inner")
print(merged_assmt.shape)
merged_assmt.head()
```

```python
merged_assmt["code_module_presentation"]=merged_assmt[['code_module','code_presentation']].agg('-'.join,axis=1)
merged_assmt.head()
```

```python
merged_assmt = merged_assmt.astype({'id_student':'string'})
merged_assmt["code_module_student_presentation"]=merged_assmt[['code_module','code_presentation','id_student']].agg('-'.join,axis=1)
merged_assmt.head()
```

```python
merged_assmt["component_score"] = merged_assmt[['score']].multiply(merged_assmt['weight'],axis='index')
merged_assmt["component_score"] = merged_assmt["component_score"].div(100)
merged_assmt.head()
```

```python
final_score = merged_assmt.groupby('code_module_student_presentation').sum()
final_score = final_score.drop(columns=['id_assessment','date_submitted','is_banked','score','date','weight'])
final_score = final_score.rename(columns = {'component_score': 'final_score'}, inplace = False)
print(final_score.shape)
final_score.head()
```

```python
integrated = integrated.merge(final_score,on="code_module_student_presentation", how="outer")
print(integrated.shape)
integrated.head()
```

```python
vle = pd.read_csv('vle.csv')
print(vle.shape)
```

```python
vle = vle.astype({'id_site':'string'})
vle["code_module_site_presentation"]=vle[['code_module','code_presentation','id_site']].agg('-'.join,axis=1)
vle = vle.drop(columns=['code_module','code_presentation','id_site'])
vle.head()
```

```python
std_vle = pd.read_csv('studentVle.csv')
print(std_vle.shape)
std_vle.head()
```

```python
std_vle = std_vle.astype({'id_site':'string'})
std_vle["code_module_site_presentation"]=std_vle[['code_module','code_presentation','id_site']].agg('-'.join,axis=1)
std_vle = std_vle.astype({'id_student':'string'})
std_vle["code_module_student_presentation"]=std_vle[['code_module','code_presentation','id_student']].agg('-'.join,axis=1)
std_vle.head()
```

```python
vle = vle.merge(std_vle,on="code_module_site_presentation", how="outer")
print(vle.shape)
vle.head()
```

```python
sum_click = vle.groupby('code_module_student_presentation').sum()
sum_click = sum_click.drop(columns=['week_from','week_to', 'date'])
sum_click.head()
```

```python
integrated = integrated.merge(sum_click,on="code_module_student_presentation", how="outer")
print(integrated.shape)
integrated.head()
```

```python
count_click = vle.groupby('code_module_student_presentation').count()
count_click = count_click.drop(columns=['week_from','week_to', 'activity_type', 'code_module_site_presentation', 'code_module','code_presentation','id_student','id_site','date'])
count_click = count_click.rename(columns = {'sum_click': 'count_click'}, inplace = False)
count_click.head()
```

```python
integrated = integrated.merge(count_click,on="code_module_student_presentation", how="outer")
print(integrated.shape)
integrated.head()
```
